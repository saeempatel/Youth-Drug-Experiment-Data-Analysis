## 2. Methods

### 2.1 Data Collection

The data set used for this analysis can be found on Kaggle, it is a free online resource for datasets.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display
```

```python
# import data
df = pd.read_csv('youth_smoking_drug_data_10000_rows_expanded.csv')
```

### 2.2 Indexing and Filtering

In this experiment, we will only be considering teenegers (10-19) in my analysis. Also we will onlly be looking at the first 400 rows in this data set since this dataset has 10000 rows.

```python
# Filter first 400 rows and age group between 10-19
df_400_youth = df.head(400)[df.head(400)['Age_Group'].isin(['10-14', '15-19'])]
```

# Analyses

### Drug Experimentation vs Media Influence

The first analysis will display the correlation between Drug Experimentation vs Media Influence.

```python
# Create the figure for the first plot
plt.figure(figsize=(6, 6))

# Drug Experimentation vs Media Influence
sns.scatterplot(data=df_400_youth, x='Media_Influence', y='Drug_Experimentation', alpha=0.6)
plt.title('Drug Experimentation vs Media Influence\nCorrelation')
plt.xlabel('Media Influence Score')
plt.ylabel('Drug Experimentation Score')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Calculate and display the average drug experimentation by Media Influence
avg_media_influence = df_400_youth.groupby('Media_Influence')['Drug_Experimentation'].agg(['mean', 'count']).round(2)
print("Average Drug Experimentation by Media Influence levels:")
print(avg_media_influence)
```

```python
# Calculate and display the correlation between Drug Experimentation and Media Influence
correlation_media_influence = df_400_youth['Drug_Experimentation'].corr(df_400_youth['Media_Influence']) * 100
print(f"Correlation between Drug Experimentation and Media Influence: {correlation_media_influence:.2f}%")
```

### Drug Experimentation vs Family Background

The second analysis will display the correlation between Drug Experimentation vs Family Background.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create the figure for the second plot
plt.figure(figsize=(6, 6))

# Drug Experimentation vs Family Background
sns.scatterplot(data=df_400_youth, x='Family_Background', y='Drug_Experimentation', alpha=0.6)
plt.title('Drug Experimentation vs Family Background\nCorrelation')
plt.xlabel('Family Background Score')
plt.ylabel('Drug Experimentation Score')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Calculate and display the average drug experimentation by Family Background
avg_family_background = df_400_youth.groupby('Family_Background')['Drug_Experimentation'].agg(['mean', 'count']).round(2)
print("Average Drug Experimentation by Family Background levels:")
print(avg_family_background)
```

```python
# Calculate and display the correlation between Drug Experimentation and Family Background
correlation_family_background = df_400_youth['Drug_Experimentation'].corr(df_400_youth['Family_Background']) * 100
print(f"Correlation between Drug Experimentation and Family Background: {correlation_family_background:.2f}%")
```

### Drug Experimentation vs Peer Influence

The third analysis will display the correlation between Drug Experimentation vs Peer Influence.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create the figure for the third plot
plt.figure(figsize=(6, 6))

# Drug Experimentation vs Peer Influence
sns.scatterplot(data=df_400_youth, x='Peer_Influence', y='Drug_Experimentation', alpha=0.6)
plt.title('Drug Experimentation vs Peer Influence\nCorrelation: 9.92%')
plt.xlabel('Peer Influence Score')
plt.ylabel('Drug Experimentation Score')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Calculate and display the average drug experimentation by Peer Influence
avg_peer_influence = df_400_youth.groupby('Peer_Influence')['Drug_Experimentation'].agg(['mean', 'count']).round(2)
print("Average Drug Experimentation by Peer Influence levels:")
print(avg_peer_influence)
```

```python
# Calculate the correlation between Drug Experimentation and Peer Influence
correlation_peer_influence = abs(df_400_youth['Drug_Experimentation'].corr(df_400_youth['Peer_Influence'])) * 100

# Print the non-negative correlation as a percentage
print(f"Correlation between Drug Experimentation and Peer Influence: {correlation_peer_influence:.2f}%")

```
